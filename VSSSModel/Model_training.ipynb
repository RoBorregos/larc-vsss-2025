```python
from ultralytics import YOLO

#Load a YOLOv8 model (yolov8n, s, l or x)
model = YOLO("yolov8s.pt")

#Use the model
results = model.train(data="/home/daniela/Desktop/VSSS/larc-vsss-2025/dataset-generator/Resources/finalOutput/data.yaml",
                      epochs=50,
                      imgsz=640,
                      batch=4,
                      patience=10,
                      name='custom_VSSS_model'
)
```

```python
from ultralytics import YOLO
import cv2
```

```python
#resume training, used for when training get cut off due to any reason, or if you
#want to resume training the next day
#model.train(resume=True)

#Load the trained model
model = YOLO('/home/daniela/Desktop/VSSS/larc-vsss-2025/VSSSModel/runs/detect/custom_VSSS_model/weights/best.pt')

#validate the model
model.val()
```

```python
model = YOLO('/home/daniela/Desktop/VSSS/larc-vsss-2025/VSSSModel/runs/detect/custom_VSSS_model/weights/best.pt')
results = model('/home/daniela/Desktop/VSSS/larc-vsss-2025/dataset-generator/Resources/finalOutput/valid/images/8.jpg')
annotated_img = results[0].plot()

cv2.imshow("YOLO inference", annotated_img)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```
