```python
from ultralytics import YOLO
```

```python


#Load a YOLOv8 model (yolov8n, s, l or x)
model = YOLO("yolov8m.pt")

#Use the model"/home/daniela/Desktop/VSSS/larc-vsss-2025/dataset-generator/Resources/finalOutput/data.yaml"
results = model.train(
                    data="/home/daniela/Desktop/VSSS/larc-vsss-2025/dataset-generator/Resources/finalOutput/data.yaml",
                    epochs=150,
                    imgsz=640,
                    batch=8,
                    perspective=0.0005,
                    scale=0.4,
                    translate=0.04,
                    degrees=5,
                    shear=1,
                    hsv_s=0.5,
                    hsv_v=0.3,
                    flipud=0.2,
                    patience=15,
                    save_period=20,
                    name="custom-yolov8m",
)
```

```python
model = YOLO('/home/daniela/Desktop/VSSS/larc-vsss-2025/VSSS_modelM/runs/detect/custom-yolov8m/weights/last.pt')
#continue training from where it left off TRAIN FROM LEFT OFF
model.train(resume=True)
```
